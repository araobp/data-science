# Using matplotlib colormap in Blender

Date: 2023/10/12

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Get colormap data from matplotlib

```python
cmap = plt.get_cmap('viridis')
cmap.N
```

```python
# Generate 256 materials in Blender with this array
rgba_array = np.array([cmap(i) for i in range(cmap.N)])
rgba_array.shape
```

```python
plt.imshow(rgba_array.reshape(1, 256, 4))
plt.axis('off')
plt.show()
```

## Create "Art of MFSCs" with Blender like this picture

```python
import glob
feature_files = glob.glob('../OscilloscopeGUI/dataset/Y2019/data_restaurant/data/*')
feature_file = feature_files[0]
```

```python
feature = np.loadtxt(feature_file, delimiter=',')
feature.shape
```

```python
mfscs = feature.reshape(400, 40)[:200,:]
mfscs
```

```python
fig, ax = plt.subplots()
img = ax.imshow(mfscs.T, cmap='viridis')
ax.invert_yaxis()
ax.set_axis_off()
```
