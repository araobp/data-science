```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df_1_440 = pd.read_csv("config1_440.txt", delimiter='\t', skiprows=2)
df_1_440_ac_off = pd.read_csv("config1_440_ac_off.txt", delimiter='\t', skiprows=2)
df_1_13000 = pd.read_csv("config1_13000.txt", delimiter='\t', skiprows=2)
df_2_13000 = pd.read_csv("config2_13000.txt", delimiter='\t', skiprows=2)
df_3_13000 = pd.read_csv("config3_13000.txt", delimiter='\t', skiprows=2)
df_4_13000 = pd.read_csv("config4_13000.txt", delimiter='\t', skiprows=2)
df_5_13000 = pd.read_csv("config5_13000.txt", delimiter='\t', skiprows=2)

df_1_440
```

```python
def plot_spectrogram(df, title):
    fig, axes = plt.subplots(1, 2, figsize=(10,2.5))
    
    ax = axes[0]
    ax.plot(df.Freq, df.Mag)
    ax.set_xlabel('Frequency(Hz)')
    ax.set_ylabel('Amplitude')
    ax.set_title('Spectrogram')
    
    ax = axes[1]
    ax.plot(df.Freq, df['Mag(dB)'])
    ax.set_xlabel('Frequency(Hz)')
    ax.set_ylabel('Magnitude(dB)')
    ax.set_title('Spectrogram')

    fig.suptitle(title, y=1.08)
    plt.show()
```

```python
plot_spectrogram(df_1_440, 'Config1, 440Hz with noise, sinc3')
```

```python
plot_spectrogram(df_1_440_ac_off, 'Config1, AC coupling off, 440Hz with noise, sinc3')
```

```python
plot_spectrogram(df_2_13000, 'Config2, 13000Hz with noise, sinc3')
```

```python
plot_spectrogram(df_3_13000, 'Config3, 13000Hz with noise, sinc3')
```

```python
plot_spectrogram(df_1_13000, 'Config1, 13000Hz with noise, sinc3')
```

```python
plot_spectrogram(df_5_13000, 'Config5, 13000Hz with noise, sinc4')
```
