# Audio Processing Basics 1

```python
import numpy as np
from scipy.io.wavfile import read, write
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt

%matplotlib inline
```

## Test audio file

"piano2.wav" from https://www.kozco.com/tech/soundtests.html

```python
from IPython.display import Audio
Audio('audio/piano2.wav')
```

```python
fs, data = read('audio/piano2.wav')
'Sampling frequency: {} Hz'.format(fs)
```

```python
data.shape  # left and right channels
```

```python
## Wave form
```

```python
#num_samples = data.shape[0]
START = 16800
END = 40000
N = END - START
T = 1 / fs
data_l = data[START:END, 0]
data_r = data[START:END, 1]
time = np.linspace(0., T, N)
fig, ax = plt.subplots(figsize=(12,2))
ax.plot(time, data_l, label='Left')
ax.plot(time, data_r, label='Right')
ax.legend()
ax.set_title('Wave form')
ax.set_xlabel('Time(sec)')
ax.set_ylabel('Amplitude')
```

## FFT

```python
yf = fft(data_l)
xf = fftfreq(N, T)[:(N)//2]
```

```python
Nyquist = fs/2
UPPER_FREQ = 880  # Hz

UPPER = int(xf.shape[0] * UPPER_FREQ / Nyquist)

fig, ax = plt.subplots(figsize=(12,2))
ax.bar(xf[1:UPPER], (2.0/N * np.abs(yf[0:N//2]))[1:UPPER], width=1)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Amplitude')
ax.set_title('FFT')
ax.grid()
```

Harmonic of musical note "G1(49Hz)" can be observed in the chart.
