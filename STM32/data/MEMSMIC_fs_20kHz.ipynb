```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df_440 = pd.read_csv("MEMSMIC_fs20kHz_AC_COUPLING_440Hz_20230918.txt", delimiter='\t', skiprows=1)
df_13000 = pd.read_csv("MEMSMIC_fs20kHz_AC_COUPLING_13000Hz_20230918.txt", delimiter='\t', skiprows=1)
df_off = pd.read_csv("MEMSMIC_fs20kHz_AC_COUPLING_OFF_20230918.txt", delimiter='\t', skiprows=1)
```

```python
def plot_spectrogram(df, title):
    fig, axes = plt.subplots(1, 2, figsize=(10,2.5))
    
    ax = axes[0]
    ax.plot(df.Freq, df.Mag)
    ax.set_xlabel('Frequency(Hz)')
    ax.set_ylabel('Amplitude')
    ax.set_title('Spectrogram')
    
    ax = axes[1]
    ax.plot(df.Freq, df['Mag(dB)'])
    ax.set_xlabel('Frequency(Hz)')
    ax.set_ylabel('Magnitude(dB)')
    ax.set_title('Spectrogram')

    fig.suptitle(title, y=1.08)
    plt.show()
```

```python
plot_spectrogram(df_440, '440kHz tone with background noise')
```

```python
# Note that aliasing is observed in the following chart
plot_spectrogram(df_13000, '13000kHz tone with background noise')
```

```python
# Note that DC is dominant in the following chart
plot_spectrogram(df_off, 'AC coupling off with background noise')
```
