```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv("teraterm_MEMSMIC_20230917.log", delimiter='\t', skiprows=4)
df
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(df.Freq, df.Mag)
ax.set_xlabel('Frequency(Hz)')
ax.set_ylabel('Amplitude')
ax.set_title('Spectrogram')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(df.Freq, df['Mag(dB)'])
ax.set_xlabel('Frequency(Hz)')
ax.set_ylabel('Magnitude(dB)')
ax.set_title('Spectrogram')
plt.show()
```
